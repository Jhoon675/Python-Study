```python
import bs4
source = driver.page_source
bs = bs4.BeautifulSoup(source,'lxml')
entire=bs.find('ul',class_='quickSearchResultBoxSidoGugun')
li_list=entire.find_all('li')
```

```python
a = entire.find_all('li',{'class':'quickResultLstCon'})
lat_data=[]
for i in range(len(a)):
    lat_data.append(a[i]['data-lat'])
long_data=[]
for i in range(len(a)):
    long_data.append(a[i]['data-long'])
```

```python
name_data =[]
for infor in li_list:
    name_data.append(infor.find('p').text)
# long_data
```

```python
# df = pd.DataFrame([x for x in zip(name_data,zip(lat_data,long_data))],columns=['지명','좌표'])
df = pd.DataFrame([x for x in zip(name_data,lat_data)],columns=['지명','좌표'])
df2 = pd.DataFrame(long_data,columns=['좌표2'])
df3 = pd.concat([df,df2],axis = 1)
c = df3['지명'].tolist()
# c[:10]
```

```python
b =[]
for i in range(len(c)):
    b.append(c[i].split(' '))
#   b는 지명 데이터를 쪼갠 것
# b
name=[]
# 구 목록을 뽑아준다,
for i in range(len(b)):
    name.append(b[i][1])
```

```python
# name
```

```python
df3['구'] = name
# df3
```

```python
a=df3['구'].value_counts()
```

```python
gu_name=a.index.tolist()
```

```python
gu_counts=a.tolist()
```

```python
df_f = pd.DataFrame([x for x in zip(gu_name,gu_counts)],columns=['name','value'])
# df_f
```

```python
from folium import Choropleth

folium.Choropleth(
    geo_data = geo_json,
    name = 'choropleth',
    data = df_f,
    columns = ['name','value'],
    key_on = 'feature.properties.name',
    fill_color = 'Oranges',
    fill_opacity = 0.85,
    line_opacity = 0.2
).add_to(m)
m
```
